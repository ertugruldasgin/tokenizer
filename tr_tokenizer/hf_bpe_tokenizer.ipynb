```python
from tokenizers import Tokenizer, models, pre_tokenizers, trainers
```

```python
spl_tokens = []
```

```python
tokenizer = Tokenizer(models.BPE(unk_token='<unk>'))
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()
```

```python
trainer = trainers.BpeTrainer(special_tokens=spl_tokens, vocab_size=20000, limit_alphabet=1200, min_frequency=16)
```

```python
from datasets import load_dataset

ds = load_dataset("wikimedia/wikipedia", "20231101.tr")
df = ds['train'].to_pandas()
df
```

```python
texts_with_title = df['title'] + " " + df['text']
texts = texts_with_title.to_list()
texts = [t.lower() for t in texts]
```

```python
texts2 = []

ds2 = load_dataset("habanoz/news-tr-1.8M")
df2 = ds2['train'].to_pandas()
df2
```

```python
texts2 = df2["text"].to_list()
texts2 = [t.lower() for t in texts2]
```

```python
total_text = texts + texts2
tokenizer.train_from_iterator(total_text, trainer=trainer)
```

```python
tokenizer.save("custom_bpe_tokenizer_v02.json")
tokenizer.get_vocab_size()
```
