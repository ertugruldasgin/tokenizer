```python
import json

with open("../turkish_tokenizer/bpe_v06.json", "r", encoding="utf-8") as f:
    bpe = json.load(f)

with open("../turkish_tokenizer/ekler_v05.json", "r", encoding="utf-8") as f:
    suffixes = json.load(f)

with open("../turkish_tokenizer/kokler_v08.json", "r", encoding="utf-8") as f:
    roots = json.load(f)

bpe_keys = set(bpe.keys())
suffixes_keys = set(suffixes.keys())
roots_keys = set(roots.keys())
```

```python
from tokenizers import Tokenizer

tokenizer = Tokenizer.from_file("custom_bpe_tokenizer_v02.json")
new_vocab = tokenizer.get_vocab()
new_tokens = set(new_vocab.keys())
```

```python
old_keys = roots_keys.union(suffixes_keys.union(bpe_keys))

intersect = new_tokens.intersection(old_keys)
different_keys = new_tokens - intersect
print(len(different_keys))
```

```python
new_diff = []
count = 0
for tok in different_keys:
    if len(tok) >= 3:
        new_diff.append(tok)

print(new_diff)
```

```python
BPE_START = 22569
BPE_END = 32767
current_id = BPE_START

old_vocab_keys = list(bpe.keys())

new_bpe = {}
i = 0
while (old_vocab_keys[i][:4] != "bpe_" and i < BPE_END - BPE_START):
    new_bpe[old_vocab_keys[i]] = current_id
    current_id += 1
    i += 1

k = 0
for j in range(current_id, BPE_END + 1):
    new_bpe[new_diff[k]] = j
    k += 1

with open("bpe_v07.json", "w", encoding="utf-8") as f:
    json.dump(new_bpe, f, ensure_ascii=False, indent=2)
```
